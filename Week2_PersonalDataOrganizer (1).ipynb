```python
# Task 3: Supplier Status Function
# Question 1 & 2:Create a function called check_supplier_status that:Takes supplier_name and on_time_rate as parameters
#and Returns a status based on the on_time_rate

def check_supplier_status(supplier_name, on_time_rate):
    """
    This function checks a supplier's status based on their on-time delivery rate.
    Parameters:
        supplier_name (str): The name of the supplier
        on_time_rate (float): The supplier's on-time delivery rate (0 to 100)
    Returns:
        str: A message with the supplier's status
    """
    if on_time_rate >= 90:
        status = "Excellent"
    elif on_time_rate >= 75:
        status = "Good"
    elif on_time_rate >= 50:
        status = "Average"
    else:
        status = "Poor"
    
    return f"Supplier {supplier_name} has a status of: {status}"

# Question 3: Test the function with at least 3 suppliers
print(check_supplier_status("A", 92))   # Excellent
print(check_supplier_status("B", 78))   # Good
print(check_supplier_status("C", 45))   # Poor
```

```python
#Task 5:Explore the dataset using for loop and logic

import csv

# Step 1: Open the CSV file
with open('supply_chain.csv', 'r') as file:
    reader = csv.reader(file)
    lines = [row for row in reader if row]  # read all rows, skip empty ones

    # Step 2: Separate header and data rows
    headers = lines[0]
    data_rows = lines[1:]

    # Question 1: How many total rows of data are in the file?
    total_rows = len(data_rows)

    # Question 2: How many unique suppliers are there?
    suppliers = set()
    for row in data_rows:
        suppliers.add(row[3])  # supplier_name is column index 3
    unique_suppliers = len(suppliers)

    # Question 3: What is the highest order cost?
    costs = [float(row[4]) for row in data_rows]  # cost is column index 4
    highest_cost = max(costs)

    # Question 4: What is the lowest order cost?
    lowest_cost = min(costs)

    # Question 5: How many orders have a specific status (e.g., "Delivered")?
    delivered_orders = sum(1 for row in data_rows if row[6] == "Delivered")  # status is column index 6

    # Step 3: Print results clearly labeled
    print("Total rows of data:", total_rows)
    print("Unique suppliers:", unique_suppliers)
    print("Highest order cost:", highest_cost)
    print("Lowest order cost:", lowest_cost)
    print("Number of 'Delivered' orders:", delivered_orders)


```

```python
#Challenge 1
# Challenge 1: Approval Function with Reliability
#Write a function that takes an order value and supplier reliability score, then returns 
# different approval levels based on both factors.

def approval_check(order_value, reliability_score):
    """
    Decide approval level based on order value and supplier reliability.
    - order_value: numeric cost of the order
    - reliability_score: supplier reliability percentage (0–100)
    """
    if order_value > 10000 and reliability_score < 70:
        return "High Risk – Requires Senior Manager Approval"
    elif order_value > 10000 and reliability_score >= 70:
        return "Requires Manager Approval"
    elif order_value <= 10000 and reliability_score < 50:
        return "Monitor Supplier – Approval with Caution"
    else:
        return "Approved Automatically"

# Test with examples
print(approval_check(12000, 65))  # High Risk
print(approval_check(15000, 85))  # Manager Approval
print(approval_check(8000, 40))   # Monitor Supplier
print(approval_check(5000, 90))   # Approved Automatically
```

```python
#challenge 2:Orders from a Specific Supplier
#Loop through the dataset and find all orders from a specific supplier. Count how many and calculate their total value.
import csv

supplier_to_check = "Alpha Supplies"  # change this to any supplier name

with open('supply_chain.csv', 'r') as file:
    reader = csv.reader(file)
    lines = [row for row in reader if row]

    headers = lines[0]
    data_rows = lines[1:]

    count_orders = 0
    total_value = 0

    for row in data_rows:
        supplier_name = row[3]
        cost = float(row[4])

        if supplier_name == supplier_to_check:
            count_orders += 1
            total_value += cost

    print(f"Supplier: {supplier_to_check}")
    print(f"Total Orders: {count_orders}")
    print(f"Total Value: {total_value}")
```

```python
# Challenge 3: Identify missing values in the supply chain dataset

# Step 1: Import the csv module to work with CSV files
import csv

# Step 2: Open the CSV file in read mode
# Make sure the file 'supply_chain.csv' exists in the same directory as this script
with open('supply_chain.csv', 'r') as file:
    
    # Step 3: Create a CSV reader object
    reader = csv.reader(file)
    
    # Step 4: Read all rows into a list, skipping any completely empty rows
    lines = [row for row in reader if row]

    # Step 5: Separate the header row (column names) from the data rows
    headers = lines[0]       # First row contains column titles
    data_rows = lines[1:]    # Remaining rows contain actual data

    # Step 6: Loop through each data row to check for missing values
    # Use enumerate to track row numbers (start=2 because row 1 is the header)
    for i, row in enumerate(data_rows, start=2):
        
        # Step 7: Create a list to store names of columns with missing values
        missing_columns = []
        
        # Step 8: Loop through each value in the row
        for j, value in enumerate(row):
            # If the value is empty (after stripping spaces), mark the column as missing
            if value.strip() == "":
                missing_columns.append(headers[j])
        
        # Step 9: If any missing values were found, print the row number and column names
        if missing_columns:
            print(f"Row {i} has missing data in columns: {', '.join(missing_columns)}")

```
